```python
from numpy import asarray
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta # Distribution type
```

# BETA Distribution

```python
a, b = 2, 7
#mean, var, skew, kurt = beta.stats(a, b, moments='mvsk')
x = np.linspace(beta.ppf(0.01, a, b),beta.ppf(0.99, a, b), 100)
vals = beta.pdf(x, a, b)

a, b = 2, 10
#mean, var, skew, kurt = beta.stats(a, b, moments='mvsk')
x_2 = np.linspace(beta.ppf(0.01, a, b),beta.ppf(0.99, a, b), 100)
vals_2 = beta.pdf(x_2, a, b)

a, b = 1, 10
#mean, var, skew, kurt = beta.stats(a, b, moments='mvsk')
x_3 = np.linspace(beta.ppf(0.01, a, b),beta.ppf(0.99, a, b), 100)
vals_3 = beta.pdf(x_3, a, b)

a, b = 2, 20
#mean, var, skew, kurt = beta.stats(a, b, moments='mvsk')
x_4 = np.linspace(0,1, 100)
vals_4 = beta.pdf(x_4, a, b)

plt.figure(1, figsize=(12,10))
plt.plot(x,vals,'red')
plt.plot(x_2,vals_2,'blue')
plt.plot(x_3,vals_3,'orange')
plt.plot(x_4,vals_4,'black')
plt.title("Histogram Original")
plt.show()
```

```python
# https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.beta.html
fig, ax = plt.subplots(1, 1)
ax.plot(x, beta.pdf(x, a, b))
rv = beta(a, b)
ax.plot(x, rv.pdf(x), 'k-', lw=2, label='frozen pdf')
vals = beta.ppf([0.001, 0.5, 0.999], a, b)
np.allclose([0.001, 0.5, 0.999], beta.cdf(vals, a, b))
r = beta.rvs(a, b, size=1000)
ax.hist(r, density=True, histtype='stepfilled', alpha=0.2)
ax.legend(loc='best', frameon=False)
plt.show()
```

# Identify your Data’s Distribution

```python
# https://towardsdatascience.com/identify-your-datas-distribution-d76062fc0802
# https://medium.com/the-researchers-guide/finding-the-best-distribution-that-fits-your-data-using-pythons-fitter-library-319a5a0972e9

import numpy as np
import pandas as pd
import seaborn as sns
from fitter import Fitter, get_common_distributions, get_distributions

mu, sigma = 0, 0.1 # mean and standard deviation
data = np.random.normal(mu, sigma, 10000)
print(data.shape)
f = Fitter(data, distributions= get_common_distributions())
f.fit()
f.summary()
```

# Q-Q Plot

```python
x1 = np.random.normal(1,2,1000)
x2 = np.random.normal(1,2,1000)

x1.sort()
x2.sort()

plt.scatter(x1,x2)
plt.plot([min(x1),max(x1)],[min(x1),max(x1)],color="red")
plt.xlabel("1st dataset's quantiles")
plt.ylabel("2nd dataset's quantiles")
plt.show()
```
