# Fourier and Wavelets

```python
#Main Libraries 
import numpy as np
import matplotlib.pyplot as plt
import datetime
import scipy.fftpack
#import tensorflow as tf
#from tensorflow import keras # And the tf and keras framework, thanks to Google
import scipy.io as io
import sys, os, time
import math
from scipy.fft import fft, fftfreq #Fourier
```

```python
from scipy.fft import fft, fftfreq

N = 600 # Number of sample points
T = 1.0 / 800.0 # sample spacing
x = np.linspace(0.0, N*T, N, endpoint=False)
y = np.sin(50.0 * 2.0*np.pi*x) + 0.5*np.sin(80.0 * 2.0*np.pi*x)

plt.plot(x,y)
plt.grid()
plt.show()

yf = fft(y)
xf = fftfreq(N, T)[:N//2]
plt.plot(xf, 2.0/N * np.abs(yf[0:N//2]))
plt.grid()
plt.show()
```

```python
t = np.arange(63)

data = []
for i in range(0,len(t)):
    data.append(3*math.cos( (math.pi/8) * t[i] ))


n = len(data)
spectrum = np.fft.fft(data) 
spectrum_magnitude = np.abs(spectrum)
spectrum_magnitude_normalized = np.abs(spectrum) / n
spectrum_phase = np.angle(spectrum)   
spectrum_energy_normalized = np.abs(spectrum) * np.abs(spectrum) / n
# if signal x(t) is Real then:
# |x(ω)| is an even function: |x(-ω)| = |x(ω)|  --> symmetry [1:N/2] <-> [N/2:N]
# φ(ω) is a odd function: φ(-ω) = -φ(ω)
# The average absolute value of the signal in time equals with the zero frequency value of the normalized spectrum =>
# => |avg(data)| = spectrum_magnitude_normalized[0]
# Parseval's theorem: the total energy of a signal will not change if we change the base, =>
# => so the energy will not change from time to frequency domain 


plt.figure(figsize=(8,6))
plt.plot(data, alpha=0.8, marker='o',color='blue')
plt.grid()
plt.show()

plt.figure(figsize=(8,6))
plt.plot(spectrum_magnitude, alpha=0.8, marker='o',color='blue',label="FFT data") #[0:n//2]
plt.ylabel("FFT Amplitude")
plt.xlabel("frequencies (hz)")
plt.title('FFT') 
plt.legend()
plt.grid()
plt.show()

plt.figure(figsize=(8,6))
plt.plot(spectrum_phase, alpha=0.8, marker='o',color='blue',label="FFT data") #[0:n//2]
plt.ylabel("FFT Phase")
plt.xlabel("frequencies (hz)")
plt.title('FFT') 
plt.legend()
plt.grid()
plt.show()

plt.figure(figsize=(8,6))
plt.plot(spectrum_magnitude_normalized, alpha=0.8, marker='o',color='blue',label="FFT data") #[0:n//2]
plt.ylabel("FFT Phase")
plt.xlabel("frequencies (hz)")
plt.title('FFT') 
plt.legend()
plt.grid()
plt.show()
```

```python
# Now: short-time Fourier transform: STfft
# Wavelet convolution PyWavelets
# --> compined with the function of cavitation
# color code magnitute of fft --> dark color for small value, white color for larnge for values
# take the log to compress the range (db)

import pywt # Wavelet Library 
```

```python
from scipy.signal import chirp, spectrogram

# Define signal
fs = 128.0
sampling_period = 1 / fs
t = np.linspace(0, 2, int(2 * fs))
x = chirp(t, 10, 2, 40, 'quadratic')

# Calculate continuous wavelet transform
coef, freqs = pywt.cwt(x, np.arange(1, 50), 'morl',
                       sampling_period=sampling_period)
```

```python
plt.plot(t, x)
plt.title("Linear Chirp, f(0)=6, f(10)=1")
plt.xlabel('t (sec)')
plt.show()
```

```python
# Show w.r.t. time and frequency
plt.figure(figsize=(10, 5))
plt.pcolormesh(t, freqs, coef, shading='gouraud')
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
#plt.yscale('log') # Set yscale
#plt.ylim([1, 100]) # Set ylim
plt.ylabel('Frequency (Hz)')
plt.xlabel('Time (sec)')
plt.show()
```
